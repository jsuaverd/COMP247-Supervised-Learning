```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
path = 'C:/Users/joans/Documents/Joan the software engr/4_term/Supervised Learning/Assignments (local)/Assignment4/student-por.csv'
data_joan = pd.read_csv(path, sep = ';', engine='python')
```

```python
data_joan['pass_joan'] = (data_joan['G1'] + data_joan['G2'] + data_joan['G3'] >= 35).astype(int)
```

```python
data_joan.drop(['G1', 'G2', 'G3'], axis=1, inplace=True)
```

```python
features_joan = data_joan.drop(['pass_joan'], axis =1)
```

```python
target_variable_joan= data_joan['pass_joan']
```

```python
##number of instances each class
target_variable_joan.value_counts()
```

```python
#separating numerical and categorical features

numeric_features_joan = features_joan.select_dtypes(include=['int64']).columns.tolist()
cat_features_joan = features_joan.select_dtypes(include=['object']).columns.tolist()

cat_features_joan
```

```python
#transform the data

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

transformer_joan = ColumnTransformer(
    transformers=[
        ("numeric", 'passthrough', numeric_features_joan),
        ("categorical", OneHotEncoder(), cat_features_joan)
    ])
```

```python
##create a decision tree classifier pipeline

from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline

clf_joan = DecisionTreeClassifier(criterion='entropy', max_depth=5)
pipeline_joan = Pipeline([
    ('preprocessor', transformer_joan),
    ('classifier', clf_joan)
])
```

```python
#split train and test set

from sklearn.model_selection import train_test_split
X_train_joan, X_test_joan, y_train_joan, y_test_joan = train_test_split(features_joan,target_variable_joan,test_size=0.2, random_state=94)
```

```python
##fit the data to the pipeline
pipeline_joan.fit(X_train_joan, y_train_joan)
```

```python
from sklearn.model_selection import cross_val_score

scores_joan = cross_val_score(pipeline_joan, X_train_joan, y_train_joan, cv=10, scoring='accuracy')
print("Cross-validation scores:", scores_joan)
print("Mean of cross-validation scores:", scores_joan.mean())
```

```python

import graphviz
from sklearn.tree import export_graphviz


transformed_feature_names = transformer_joan.get_feature_names_out()


dot_data = export_graphviz(clf_joan, out_file=None, feature_names=transformed_feature_names, class_names=['Fail', 'Pass'], filled=True, rounded=True)
graph = graphviz.Source(dot_data)
graph.render("decision_tree_joan")
```

```python
#print accuracy score
train_accuracy = pipeline_joan.score(X_train_joan, y_train_joan)
test_accuracy = pipeline_joan.score(X_test_joan, y_test_joan)
print("Training accuracy:", train_accuracy)
print("Testing accuracy:", test_accuracy)
```

```python
#metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix

#predict
y_pred_joan = pipeline_joan.predict(X_test_joan)

#measure accuracy
accuracy = accuracy_score(y_test_joan, y_pred_joan)
precision = precision_score(y_test_joan, y_pred_joan)
recall = recall_score(y_test_joan, y_pred_joan)
conf_matrix = confusion_matrix(y_test_joan, y_pred_joan)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("Confusion Matrix:\n", conf_matrix)
```

```python
#Fine tuning
from sklearn.model_selection import RandomizedSearchCV

parameters = {
    'classifier__min_samples_split': range(10, 300, 20),
    'classifier__max_depth': range(1, 30, 2),
    'classifier__min_samples_leaf': range(1, 15, 3)
}

grid_search_joan = RandomizedSearchCV(estimator=pipeline_joan,
                                      param_distributions=parameters,
                                      scoring='accuracy',
                                      cv=5,
                                      n_iter=7,
                                      refit=True,
                                      verbose=3,
                                      )
```

```python
grid_search_joan.fit(X_train_joan, y_train_joan)
```

```python
#Step 22: Print best parameters
print("Best parameters:", grid_search_joan.best_params_)

# Step 23: Print score of the model
print("Grid search score:", grid_search_joan.best_score_)
# Compare with original score from Step 14 and explain differences

# Step 24: Print best estimator
best_estimator_joan = grid_search_joan.best_estimator_
print("Best estimator:", best_estimator_joan)
```

```python

# Step 25: Fit test data using best estimator
best_estimator_joan.fit(X_train_joan, y_train_joan)

```

```python
# Step 26: Print precision, recall, and accuracy
y_pred_joan_tuned = best_estimator_joan.predict(X_test_joan)
precision_tuned = precision_score(y_test_joan, y_pred_joan_tuned)
recall_tuned = recall_score(y_test_joan, y_pred_joan_tuned)
accuracy_tuned = accuracy_score(y_test_joan, y_pred_joan_tuned)
print("Tuned Model Precision:", precision_tuned)
print("Tuned Model Recall:", recall_tuned)
print("Tuned Model Accuracy:", accuracy_tuned)
```

```python
# Step 27: Save model using joblib
import joblib

joblib.dump(best_estimator_joan, 'decision_tree_model_joan.pkl')

# Step 28: Save pipeline using joblib
joblib.dump(pipeline_joan, 'pipeline_joan.pkl')
```
